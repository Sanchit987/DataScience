```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("cost_revenue_clean.csv")
```

```python
df.describe()
```

```python
X = pd.DataFrame(df,columns=['production_budget_usd'])
Y = pd.DataFrame(df,columns=['worldwide_gross_usd'])
```

```python
plt.scatter(X,Y,alpha=0.3)
plt.title("production cost vs worldwide gross")
plt.xlabel("production cost")
plt.ylabel("worldwide gross")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.figure(figsize=(10,6))
plt.show()
```

```python
regression = LinearRegression()
```

```python
regression.fit(X ,Y)
```

```python
regression.coef_ #slope of the line in simple language
```

```python
regression.intercept_ 
```

```python
plt.scatter(X,Y,alpha=0.3)
plt.title("production cost vs worldwide gross")
plt.xlabel("production cost")
plt.ylabel("worldwide gross")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.plot(X, regression.predict(X),color = "orange",linewidth = 3)
plt.figure(figsize=(10,6))
plt.show()
```
