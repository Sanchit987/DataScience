## Notebook Imports and Packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# $ Example 1 - A simple Cost Function $

## $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
#makeing data
x_1 = np.linspace(start = -3,stop = 3,num = 500)
```

```python
#ploting function
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)
plt.plot(x_1,f(x_1))
plt.style.use('ggplot')
plt.show()
```

# slope and derivative

```python
def df(x):
    return 2*x + 1
```

```python
#making data
x_2 = np.linspace(start = -3,stop = 3,num = 500)
```

```python
#plotting derivative

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.plot(x_1,df(x_1))
plt.style.use('ggplot')
plt.show()
```

```python
#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_1,f(x_1))
plt.style.use('ggplot')

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_1,df(x_1))
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)

plt.show()
```
