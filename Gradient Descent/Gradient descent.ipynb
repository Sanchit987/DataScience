```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# $ Example 1 - A simple Cost Function $

## $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
#makeing data
x_1 = np.linspace(start = -3,stop = 3,num = 500)
```

```python
#ploting function
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)
plt.plot(x_1,f(x_1))
plt.style.use('ggplot')
plt.show()
```

# slope and derivative

```python
def df(x):
    return 2*x + 1
```

```python
#making data
x_2 = np.linspace(start = -3,stop = 3,num = 500)
```

```python
#plotting derivative

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.plot(x_1,df(x_1))
plt.style.use('ggplot')
plt.show()
```

```python
#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_1,f(x_1))
plt.style.use('ggplot')

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_1,df(x_1),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)

plt.show()
```

# Gradient Descent

```python
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.000001

x_list = [new_x]
y_list = [df(new_x)]
for i in range(500):
    prev_x = new_x
    gradient = df(new_x)
    new_x = prev_x - step_multiplier * gradient
    
    x_list.append(new_x)
    y_list.append(df(new_x))
    
    if(abs(new_x - prev_x) < precision):
        print("Looping iterations : ", i)
        break
        
        
print("Current value of Gradient : ",gradient)
print("Current minima : ",df(new_x))
print("Current f(x) : ",f(new_x))
```

```python
#Gradient descent superimposed together with previous graph


plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

plt.plot(x_1,f(x_1),linewidth = 5,alpha = 0.6)
values = np.array(x_list)
plt.scatter(values,f(values),color = "blue",s=100)
plt.style.use('ggplot')


plt.subplot(1,2,2)
plt.plot(x_1,df(x_1),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
slope_list = np.array(y_list)
plt.scatter(values,slope_list,color = "red",s = 100,alpha = 0.6)
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.show()
```
