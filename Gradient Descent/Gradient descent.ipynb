# Gradient Descent

```python
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.000001

x_list = [new_x]
y_list = [df(new_x)]
for i in range(500):
    prev_x = new_x
    gradient = df(new_x)
    new_x = prev_x - step_multiplier * gradient
    
    x_list.append(new_x)
    y_list.append(df(new_x))
    
    if(abs(new_x - prev_x) < precision):
        print("Looping iterations : ", i)
        break
        
        
print("Current value of Gradient : ",gradient)
print("Current minima : ",df(new_x))
print("Current f(x) : ",f(new_x))
```

```python
#Gradient descent superimposed together with previous graph


plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

plt.plot(x_1,f(x_1),linewidth = 5,alpha = 0.6)
values = np.array(x_list)
plt.scatter(values,f(values),color = "blue",s=100)
plt.style.use('ggplot')


plt.subplot(1,2,2)
plt.plot(x_1,df(x_1),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
slope_list = np.array(y_list)
plt.scatter(values,slope_list,color = "red",s = 100,alpha = 0.6)
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.show()
```

# Gradient Descent   as a  python Function

```python
def gradient_descent(derivative_func,initial_guess,step_multiplier=0.02,precision=0.001,max_iter = 300):
    new_x = initial_guess

    x_list = [new_x]
    y_list = [derivative_func(new_x)]
    
    for i in range(max_iter):
        prev_x = new_x
        gradient = derivative_func(prev_x)
        new_x = prev_x - step_multiplier * gradient

        x_list.append(new_x)
        y_list.append(derivative_func(new_x))

        if(abs(new_x - prev_x) < precision):
            print("Looping iterations : ", i)
            break
        
    return new_x, x_list, y_list
```

# Batch gradient descent with Sympy

```python
#setup
step_multiplier = 0.1
max_iter = 500
params = np.array([1.8,1.0]) #initial guess

for n in range(max_iter):
    gradient_x = diff(f(a,b), a).evalf(subs = {a:params[0],b:params[1]})
    gradient_y = diff(f(a,b), b).evalf(subs = {a:params[0],b:params[1]})
    gradients = np.array([gradient_x,gradient_y])
    
    params = params - step_multiplier * gradients
    
#result
print("value in gradient arrays :-" )
print("minimum at x" , params[0])
print("minimum at y", params[1])
print("evaluation : ", f(params[0],params[1]))
```

# Partial derivatives and Sybolic computation

## $$ \frac{\partial f}{\partial x} = \frac{2x \ln(3) \cdot 3^{-x^2 -y^2}}{\left( 3^{-x^2-y^2} + 1 \right) ^2}  $$
## $$ \frac{\partial f}{\partial y} = \frac{2y \ln(3) \cdot 3^{-x^2 -y^2}}{\left( 3^{-x^2-y^2} + 1 \right) ^2}  $$

```python
#partial derivative function

def fpx(x,y):
    r = 3**(-x**2-y**2)
    return ( 2*x*math.log(3)*r )/(r+1)**2 

def fpy(x,y):
    r = 3**(-x**2-y**2)
    return ( 2*y*math.log(3)*r )/(r+1)**2
```

```python
#setup
step_multiplier = 0.1
max_iter = 500
params = np.array([1.8,1.0]) #initial guess

for n in range(max_iter):
    gradient_x = fpx(params[0],params[1])
    gradient_y = fpy(params[0],params[1])
    gradients = np.array([gradient_x,gradient_y])
    
    params = params - step_multiplier * gradients
    
#result
print("value in gradient arrays :-" )
print("minimum at x" , params[0])
print("minimum at y", params[1])
print("evaluation : ", f(params[0],params[1]))
```

## Mean Squared error : a cost function for regression problems
### $$ RSS = \sum_{i=1}^{n}\big( y^{(i)} - h_\theta x^{(i)} \big)^2$$
### $$ MSE = \frac{1}{n} \sum_{i=1}^{n}\big( y^{(i)} - h_\theta x^{(i)} \big)^2$$

```python
#make sampled data
x_5 = np.array([1.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]).reshape(7,1)
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7,1)
```

```python
#MSE
def mse(y,y_hat):
    sum_of_terms = 0
    n = len(y)
    for i in range(n):
        sum_of_terms += (y-y_hat)**2
    return (1/n)*sum_of_terms
```

# Partial Derivatives wrt $\theta_0$ & $\theta1$

## $$\frac{\partial MSE}{\partial \theta_0} = -\frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_0} = -\frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size
    
    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)
    
    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)
    
    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))
    
# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
