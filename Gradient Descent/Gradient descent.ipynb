# $ Example 1 - A simple Cost Function $

## $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
#makeing data
x_1 = np.linspace(start = -3,stop = 3,num = 500)
```

```python
#ploting function
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)
plt.plot(x_1,f(x_1))
plt.style.use('ggplot')
plt.show()
```

# slope and derivative

```python
def df(x):
    return 2*x + 1
```

```python
#making data
x_2 = np.linspace(start = -3,stop = 3,num = 500)
```

```python
#plotting derivative

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.plot(x_1,df(x_1))
plt.style.use('ggplot')
plt.show()
```

```python
#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_1,f(x_1))
plt.style.use('ggplot')

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_1,df(x_1),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)

plt.show()
```

# Gradient Descent

```python
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.000001

x_list = [new_x]
y_list = [df(new_x)]
for i in range(500):
    prev_x = new_x
    gradient = df(new_x)
    new_x = prev_x - step_multiplier * gradient
    
    x_list.append(new_x)
    y_list.append(df(new_x))
    
    if(abs(new_x - prev_x) < precision):
        print("Looping iterations : ", i)
        break
        
        
print("Current value of Gradient : ",gradient)
print("Current minima : ",df(new_x))
print("Current f(x) : ",f(new_x))
```

```python
#Gradient descent superimposed together with previous graph


plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

plt.plot(x_1,f(x_1),linewidth = 5,alpha = 0.6)
values = np.array(x_list)
plt.scatter(values,f(values),color = "blue",s=100)
plt.style.use('ggplot')


plt.subplot(1,2,2)
plt.plot(x_1,df(x_1),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
slope_list = np.array(y_list)
plt.scatter(values,slope_list,color = "red",s = 100,alpha = 0.6)
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.show()
```

# $ Example 2 - Multiple Minima Function $

## $f(x) = x^4 + 4x^2 + 5$

```python
#makeing some data
x_2 = np.linspace(-2,2,1000)

def g(x):
    return x**4 - 4*(x**2) + 5

def dg(x):
    return 4*(x**3) - 8*x
```

```python
#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize = 16)
plt.ylabel("g(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_2,g(x_2))
plt.style.use('ggplot')

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_2,dg(x_2),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-2,2)
plt.ylim(-6,8)

plt.show()
```

# Gradient Descent   as a  python Function

```python
def gradient_descent(derivative_func,initial_guess,step_multiplier=0.02,precision=0.001):
    new_x = initial_guess

    x_list = [new_x]
    y_list = [derivative_func(new_x)]
    
    for i in range(500):
        prev_x = new_x
        gradient = derivative_func(prev_x)
        new_x = prev_x - step_multiplier * gradient

        x_list.append(new_x)
        y_list.append(derivative_func(new_x))

        if(abs(new_x - prev_x) < precision):
            print("Looping iterations : ", i)
            break
        
    return new_x, x_list, y_list
```

```python
local_min, list_x, deri_list = gradient_descent(dg,0.5,0.02,0.001)
print("Current value of Gradient : ",local_min)
```

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dg,initial_guess = -0.5)
print("Current value of Gradient : ",local_min)
```

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dg,initial_guess = 0.1)

#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize = 16)
plt.ylabel("g(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_2,g(x_2))
plt.style.use('ggplot')
plt.scatter(list_x,g(np.array(list_x)))

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_2,dg(x_2),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.scatter(list_x,deri_list)

plt.show()
```

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dg,initial_guess = -0.1)

#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize = 16)
plt.ylabel("g(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_2,g(x_2))
plt.style.use('ggplot')
plt.scatter(list_x,g(np.array(list_x)))

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_2,dg(x_2),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.scatter(list_x,deri_list)

plt.show()
```

On Examining the above two cells we come to know that there is a problem due to two minimas present because when we pass different inital values we are getting two different values which is a drawback and hence we need to correct the problem.

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dg,initial_guess = 0)

#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize = 16)
plt.ylabel("g(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_2,g(x_2))
plt.style.use('ggplot')
plt.scatter(list_x,g(np.array(list_x)))

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_2,dg(x_2),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.scatter(list_x,deri_list)

plt.show()
```

We notice here there is a problem of local minimum and global minimum and we also notice that if slope turns out to be zero at any other point we would not be able to find the minima.
