```python
def f(x):
    return x**2 + x + 1
```

# slope and derivative

```python
def df(x):
    return 2*x + 1
```

```python
#making data
x_2 = np.linspace(start = -3,stop = 3,num = 500)
```

```python
#plotting derivative

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.plot(x_1,df(x_1))
plt.style.use('ggplot')
plt.show()
```

# Gradient Descent

```python
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.000001

x_list = [new_x]
y_list = [df(new_x)]
for i in range(500):
    prev_x = new_x
    gradient = df(new_x)
    new_x = prev_x - step_multiplier * gradient
    
    x_list.append(new_x)
    y_list.append(df(new_x))
    
    if(abs(new_x - prev_x) < precision):
        print("Looping iterations : ", i)
        break
        
        
print("Current value of Gradient : ",gradient)
print("Current minima : ",df(new_x))
print("Current f(x) : ",f(new_x))
```

```python
#Gradient descent superimposed together with previous graph


plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

plt.plot(x_1,f(x_1),linewidth = 5,alpha = 0.6)
values = np.array(x_list)
plt.scatter(values,f(values),color = "blue",s=100)
plt.style.use('ggplot')


plt.subplot(1,2,2)
plt.plot(x_1,df(x_1),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
slope_list = np.array(y_list)
plt.scatter(values,slope_list,color = "red",s = 100,alpha = 0.6)
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.show()
```

# $ Example 2 - Multiple Minima Function $

## $f(x) = x^4 + 4x^2 + 5$

```python
#makeing some data
x_2 = np.linspace(-2,2,1000)

def g(x):
    return x**4 - 4*(x**2) + 5

def dg(x):
    return 4*(x**3) - 8*x
```

```python
#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize = 16)
plt.ylabel("g(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_2,g(x_2))
plt.style.use('ggplot')

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_2,dg(x_2),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-2,2)
plt.ylim(-6,8)

plt.show()
```

# Gradient Descent   as a  python Function

```python
def gradient_descent(derivative_func,initial_guess,step_multiplier=0.02,precision=0.001,max_iter = 300):
    new_x = initial_guess

    x_list = [new_x]
    y_list = [derivative_func(new_x)]
    
    for i in range(max_iter):
        prev_x = new_x
        gradient = derivative_func(prev_x)
        new_x = prev_x - step_multiplier * gradient

        x_list.append(new_x)
        y_list.append(derivative_func(new_x))

        if(abs(new_x - prev_x) < precision):
            print("Looping iterations : ", i)
            break
        
    return new_x, x_list, y_list
```

```python
local_min, list_x, deri_list = gradient_descent(dg,0.5,0.02,0.001)
print("Current value of Gradient : ",local_min)
```

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dg,initial_guess = -0.5)
print("Current value of Gradient : ",local_min)
```

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dg,initial_guess = 0.1)

#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize = 16)
plt.ylabel("g(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_2,g(x_2))
plt.style.use('ggplot')
plt.scatter(list_x,g(np.array(list_x)))

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_2,dg(x_2),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.scatter(list_x,deri_list)

plt.show()
```

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dg,initial_guess = -0.1)

#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize = 16)
plt.ylabel("g(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_2,g(x_2))
plt.style.use('ggplot')
plt.scatter(list_x,g(np.array(list_x)))

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_2,dg(x_2),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.scatter(list_x,deri_list)

plt.show()
```

On Examining the above two cells we come to know that there is a problem due to two minimas present because when we pass different inital values we are getting two different values which is a drawback and hence we need to correct the problem.

We notice here there is a problem of local minimum and global minimum and we also notice that if slope turns out to be zero at any other point we would not be able to find the minima.

# Example 3 - Divergence, OverFlow, Python Tuples

# $ h(x) = x^5 - 2x^4 + 2 $

```python
x_3 = np.linspace(start = -2.5,stop = 2.5, num = 1000)

def h(x):
    return x**5 - 2 * x**4 + 2

def dh(x):
    return 5*x**4 - 8 * x**3
```

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dh,initial_guess = 0.2)

#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-1.2,2.5)
plt.ylim(-1,4)
plt.xlabel("x",fontsize = 16)
plt.ylabel("h(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_3,h(x_3))
plt.style.use('ggplot')
plt.scatter(list_x,h(np.array(list_x)),color = "blue")

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_3,dh(x_3),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-1,2)
plt.ylim(-4,5)
plt.scatter(list_x,deri_list)

plt.show()

print("Local Min :",local_min)
print("cost at this minimum :",h(local_min))
print("number of steps : ", len(list_x))
```

```python
local_min, list_x, deri_list = gradient_descent(derivative_func = dh,initial_guess = -0.2,max_iter = 71)

#Both plots together
plt.figure(figsize = [10,3])

plt.subplot(1,2,1)

plt.xlim(-1.2,2.5)
plt.ylim(-1,4)
plt.xlabel("x",fontsize = 16)
plt.ylabel("h(x)",fontsize = 16)

#chart 1 : cost function
plt.plot(x_3,h(x_3))
plt.style.use('ggplot')
plt.scatter(list_x,h(np.array(list_x)),color = "blue")

#chart 2 : Derivative
plt.subplot(1,2,2)
plt.plot(x_3,dh(x_3),color = "blue")
plt.xlabel("x",fontsize = 16)
plt.ylabel("dg(x)",fontsize = 16)
plt.xlim(-1,2)
plt.ylim(-4,5)
plt.scatter(list_x,deri_list)

plt.show()

print("Local Min :",local_min)
print("cost at this minimum :",h(local_min))
print("number of steps : ", len(list_x))
```

Now we come to kow the importance of step multiplier i.e. it is a very important part for running optimization algorithms a small multiplier might taker very long to make run the Algorithm and a very large multiplier might not allow one to find the required minima.

# Finding an optimal learning rate

```python

n = 100
low_gamma = gradient_descent(derivative_func = dg,initial_guess = 3,
                                                step_multiplier=0.0005,precision=0.0001,max_iter = n)

mid_gamma = gradient_descent(derivative_func = dg,initial_guess = 3,
                                                step_multiplier=0.001,precision=0.0001,max_iter = n)

high_gamma = gradient_descent(derivative_func = dg,initial_guess = 3,
                                                step_multiplier=0.002,precision=0.0001,max_iter = n)

insane_gamma = gradient_descent(derivative_func = dg,initial_guess = 1.9,
                                                step_multiplier=0.25,precision=0.0001,max_iter = n)
#Plotting reduction in cost for each iteration
plt.figure(figsize = [20,5])

plt.xlim(0,n)
plt.ylim(0,50)
plt.title("Effect of learning rate")
plt.xlabel("Number of iteration",fontsize = 16)
plt.ylabel("cost",fontsize = 16)

low_values = np.array(low_gamma[1])
mid_values = np.array(mid_gamma[1])
high_values = np.array(high_gamma[1])
insane_values = np.array(insane_gamma[1])

iteration_list = list(range(0,n+1))

#chart for low_gamma (learning rate = 0.0005)
plt.plot(iteration_list,g(low_values),color = "lightgreen",linewidth = 5,alpha = 0.5,)
plt.style.use('ggplot')
plt.scatter(iteration_list,g(low_values),color = "darkgreen")

#chart for mid_gamma (learning rate = 0.001)

plt.plot(iteration_list,g(mid_values),color = "orange",linewidth = 5,alpha = 0.5)
plt.style.use('ggplot')
plt.scatter(iteration_list,g(mid_values),color = "red")

#chart for high_gamma (learning rate = 0.002)

plt.plot(iteration_list,g(high_values),color = "lightblue",linewidth = 5,alpha = 0.5)
plt.style.use('ggplot')
plt.scatter(iteration_list,g(high_values),color = "darkblue")

#chart for high_gamma (learning rate = 0.002)

plt.plot(iteration_list,g(insane_values),color = "lightpink",linewidth = 5,alpha = 0.5)
plt.style.use('ggplot')
plt.scatter(iteration_list,g(insane_values),color = "hotpink")

plt.show()
```

A decrease in cost as we increase the iterations and as the step multiplier increases first the algorithm becomes more optimal and then breaks.

# Example 4 - Data Viz with 3D charts

# Minimize $f(x,y) = \frac{1}{3^{-x^2-y^2} + 1 }$

```python
def f(x,y):
    r = 3**(-x**2 - y**2)
    return 1 / (r+1)
```

```python
# make data

x_4 = np.linspace(start = -2,stop = 2, num = 200)
y_4 = np.linspace(start = -2,stop = 2, num = 200)

x_4,y_4 = np.meshgrid(x_4,y_4)
```

```python
#Genrating 3D plot
fig = plt.figure(figsize=[16,12])
ax = fig.gca(projection = "3d")
ax.set_zlim(0.6,1)
ax.set_xlabel("x",fontsize = 10)
ax.set_ylabel("y",fontsize = 10)
ax.set_zlabel("f(x,y)-cost",fontsize = 10)
ax.plot_surface(x_4,y_4,f(x_4,y_4),cmap = cm.twilight_shifted_r,alpha = 0.4)
plt.show()
```
